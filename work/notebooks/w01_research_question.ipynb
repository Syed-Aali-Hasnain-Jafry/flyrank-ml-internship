## 2. The question: decision, action and cost of a wrong call

### Search question

**Which content pages should a content or SEO team review first for a possible refresh, based on observable performance, content-age and trend signals?**

### Unit of analysis

The unit of analysis is **one anonymised content page** represented by one row in the starter dataset.

### Decision

The decision is:

**Which pages should be placed near the top of a limited content-review queue?**

The decision is not whether a page must automatically be rewritten. It is only whether available evidence suggests that the page deserves earlier human investigation.

### Intended output

The intended output is a ranked table containing:

* the anonymised page identifier;
* an opportunity or priority score;
* the most important reason codes;
* a suggested human-review action;
* relevant supporting signals.

Possible reason codes may include:

* declining search trend;
* older content;
* substantial historical impressions;
* weak click-through performance;
* traffic opportunity;
* conflicting or uncertain signals.

### Action

A content or SEO team member could use the ranked queue to:

1. open and inspect the highest-ranked pages;
2. confirm whether the observed decline or opportunity is meaningful;
3. check whether the page is outdated, weak, duplicated or affected by another cause;
4. decide whether to refresh, optimise, consolidate, leave unchanged or investigate further;
5. record the final human decision.

### Cost of a false positive

A false positive would occur when the system ranks a page highly even though it does not require a refresh.

The possible costs include:

* wasting the reviewer’s limited time;
* changing content that was already performing appropriately;
* reducing traffic by editing a stable page unnecessarily;
* ignoring seasonality or temporary fluctuations;
* reducing confidence in the ranking system.

### Cost of a false negative

A false negative would occur when the system gives a low rank to a page that genuinely needs attention.

The possible costs include:

* allowing useful content to continue losing visibility or clicks;
* missing a recoverable traffic opportunity;
* delaying an important update;
* directing resources toward lower-value pages instead.

### Why data or machine learning may help

A content team may have more pages than it can manually inspect. Data analysis may help combine several observable signals consistently and identify pages that deserve earlier attention.

Machine learning may help only if it produces a ranking that is more useful, stable and generalisable than a clear rule-based baseline. A model is not the goal by itself. The goal is to improve the quality of the human review queue and support a better allocation of limited review time.
